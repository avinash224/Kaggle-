1 ) Implement function myBMI() that takes as input a person's height(in inches) and weight(in pounds) and computes the person's Body Mass Index(BMI). The BMI formula is:
	    
- bmi= weight * 703/(height**2)

Your function should print the string 'Underweight' if bmi<18.5, 'Normal' if 18.5<=bmi<25,and Overweight if bmi>=25.

```python
def myBMI(height_in , weights_pnd):
    bmi = weights_pnd * 703 / ((height_in)**2)
    if 18.5<=bmi<25 : 
        print('Normal')
    elif bmi > 25: 
        print('Overweight')
    else:
        print('Underweight')
```

```python
bmi(70,160)
```

```python
#using lambda function 
bmi_1 = lambda height_in , weights_pnd: (weights_pnd * 703 / ((height_in)**2)) 
body_type = lambda bmi_1 : 'Normal' if ((bmi_1 <25)and(bmi_1>=18.5))  else ('Overweight' if (bmi_1 > 25) else 'Underweight')
body_type(bmi_1(70,160))
```

2) Write a function power() that takes a positive integer n as input and prints,on the screen,all the powers of 2 from 2^1 to 2^n.

```python
def power(n):
    for i,v in enumerate(range(1,n+1)):
        print(f"for power {i} value is: {2**v}")        
```

3) Write function arithmetic() that takes a list of integers as input and returns True if they form a sequence..

```python
def sequence(list1):
    prev_val = 0
    y = 0
    for index,new_val in enumerate(list1):
        if index == 0:
            prev_val = new_val
        else:
            if prev_val + 1 == new_val:
                prev_val = new_val
                y = 1
            else:
                y=0
                break
        
    if y == 1:
        return 'list is a sequence'
    else:
        return 'list is not an sequence'         
```

```python
sequence([2,4,6])
```

4) Write function arithmetic() that takes a list of integers as input and returns True if they form an arithmetic sequence..

```python
#arithmatic sequence
def arithmatic(list1):
    y = 0
    if len(list1) < 2:
        return True 
    else:
        d = list1[0] - list1[1]
        for i in range (len(list1)-1):
            if list1[i] - list1[i+1] == d:
                y = 1
            else:
                y= 0
    
    if y == 1:
        return True 
    else:
        return False
                
```

5) Implement function factorial()that takes as input a nonnegative integer and returns its factorial.

```python
def factoral(number):
    if number >= 0:
        result = 1
        for i in range(number):
            result = result *(i+1)
        print(result)
    else:
        print("please enter number greater than 0")
```

6) An acronym is a word formed by taking the first letters of the words in a phrase and then making a word from them.Write a function acronym() that takes a phrase as input and then returns the acronym for that phrase.

```python
def acronym(word):
    x = ''
    word_list = word.split(" ")
    for i in word_list:
        x=x+i[0].upper()
    return x
```

```python
acronym('my name is')
```

7) Write function divisors() that takes a positive integer n as input and returns the list of all positive divisors of n

```python
def divisors(n):
    res = []
    for i in range(1,n+1):
        if n%i==0:
            res.append(i)
        else:
            pass
    
    return res
```

8) Write a function xmult() that takes two lists of integers as input and returns a list containing all products of integers from the list with the integers from the second list

```python
def xmult(lst1, lst2):
    result=[]
    for i in lst1:
        for j in lst2: 
            result.append(i*j)
            
    return result 
```

9) Write a function bubbleSort() that takes a list of numbers as input and sorts the list using this approach.

```python
def bubblesort(sortlist):
    new_list = []
    
    counter = len(sortlist) - 2
    while (counter!=0):
        for i in range(0, len(sortlist)-1):
            if sortlist[i] > sortlist[i+1]:
                temp=sortlist[i]
                sortlist[i]=sortlist[i+1]
                sortlist[i+1]=temp
        counter -=1
    return sortlist
            
```

11) Write a function interest() that takes one input , a floating-point interest rate.Your fuction should compute and return how long it will take for an investment to double in value.

```python
def interest(amount , rate):
        am = amount
        count=0
        while am < 2*amount:
            count=count+1
            am=am+(am*rate)
        print(count)
```

```python
interest(50, 0.06)
```
